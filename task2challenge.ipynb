````python
testing_data = testing_df[['Value']]
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
binaryclassifier = LinearDiscriminantAnalysis()
binaryclassifier.fit(training_data, training_class)
prediction = binaryclassifier.predict(testing_data)
testing_df['Class'] = prediction
testing_df
````

````python
def f1(c1):
    #Mean
    m1 = np.mean(training_data1)
    #Standard Deviation
    s1 = np.std(training_data1)
    #Normalization: Z = (X - Mean)/Std. Dev. 
    z1 = (c1-m1)/s1
    f1 = ((np.sqrt((2*np.pi)))**-1)*np.exp(-0.5*z1**2)
    return f1;

def f0(c0):
    #Mean
    m0 = np.mean(training_data0)
    #Standard Deviation
    s0 = np.std(training_data0)
    #Normalization: Z = (X - Mean)/Std. Dev. 
    z0 = (c0-m0)/s0
    f0 = ((np.sqrt((2*np.pi)))**-1)*np.exp(-0.5*z0**2)
    return f0;
````

````python
#Reading again
testing_df = pd.read_csv('task2testing.csv')
````

````python
#Condition for binary classification
for i in range(testing_df.shape[0]):
    if f1(testing_df.iloc[i, 1]) > f0(testing_df.iloc[i, 1]):
        testing_df.iloc[i,2] = 1
    else:
        testing_df.iloc[i,2] = 0
````

````python
#Writing onto a CSV file
testing_df.to_csv('task2testing.csv', index = False)
print(testing_df)
````
